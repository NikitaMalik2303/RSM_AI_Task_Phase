```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([10,20,30,40,50,60,70,80,90])
Y = np.array([15,35,30,47,60,67,70,80,95,105])
```

```python
X = X.reshape(-1,1)
Y = Y.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,Y)
```

```python
Y_pred = lr.predict(Y)
```

```python
print(Y_pred)
```

```python
plt.scatter(X, Y, c = "red")
plt.plot(X, Y_pred, c = "blue")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.show()
```

```python
lr.score(X,Y)
```
